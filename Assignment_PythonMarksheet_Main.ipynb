```python
# import functions and class from "Assignment_PythonMarksheet_Class" file
from ipynb.fs.full.Assignment_PythonMarksheet_Class import check_file_existence, ExitProgramException, ReturnToHomePageException, PythonMarksheet 

# Import os module for file operations
import os
# Import matplotlib for plot histogram
import matplotlib.pyplot as plt
# Import numpy for numerical operations
import numpy as np

        
def print_student_info(student):  # Function to print student information
    print("\nStudent ID:", student.get_student_id())
    print("Student Name:", student.get_student_name())
    print("Test 1 Mark:", student.get_test1())
    print("Assignment 1 Mark:", student.get_assignment1())
    print("Test 2 Mark:", student.get_test2())
    print("Assignment 2 Mark:", student.get_assignment2())
    print("Final Exam Mark:", student.get_final_exam())
    print("Total Mark:", student.get_total_mark())
    print("Final Grade:", student.get_final_grade())


def display_userManual():   # Function to display the user manual
    print("""
    +---------------------------------------------------------------------------------------------+
    |                                     USER MANUAL                                             |
    +---------------------------------------------------------------------------------------------+
   \t Welcome To Subject Marksheet Management System! Follow the user manual to get started: 

    << Home Menu Options >>
    1. Create New File (Option 1):
    - Create a new file to store new student records.
    - Enter a file name when prompted.
    - If the file name exists, choose a different file name.
    - The file name is case-insensitive.

    2. Load Data From Existing File (Option 2):
    - Load student records from an existing file.
    - Enter the file name with the correct format (e.g., .txt).
    - If the file is not found, re-enter a valid existing file name.
    - The file name is case-insensitive.

    3. Exit Program (Option 3)

    --  After selecting option 1 or 2, you'll enter the Student Record Management Menu  --

    << Student Record Management Menu >>
    1. Add New Student Record (Option 1):
    - Input student records and marks.
    - Choose to add more student records or return to the Student Record Management Menu.

    2. Display All Student Records (Option 2):
    - View all existing student records.

    3. Search Student Record by ID (Option 3):
    - Retrieve specific student records by the student ID.

    4. Update Student Record by ID (Option 4):
    - Update student information or marks.
    - You will get to choose which information or marks need to be modified.

    5. Delete Student Record by ID (Option 5):
    - Remove a student record by the Student ID.

    6. Save Records to File (Option 6):
    - Save all record changes to the file to permanently save the records.

    7. Generate Grade Distribution Table (Option 7):
    - Get the Grade Distribution Table according to your all current records.
    - Visualize how many students received in each grade category.

    8. Generate Grade Distribution Histogram (Option 8):
    - Visualize how many students received in each grade category.

    9. Generate Final Marks Histogram (Option 9):
    - Visualize how many students received in each final marks range (0-100).

    10. Go back to Home Page (Option 10):
    - Return to the main menu.

    11. Exit Program (Option 11)


    [ Reminder: all user inputs (e.g., file_name, student_ID, student_Name) are case-insensitive ]
    +--------------------------------------------------------------------------------------------+
    """)


    
def display_home_menu():    # Function to display the home menu
    print("""
    >>>             *  Home Menu  *          <<< 
-----------------------------------------------------------    
    1. Create A New File
    2. Open and Load Data From A Existing File
    3. Exit Program
-----------------------------------------------------------
    """)

def create_new_file(students):    # Function to create a new file
    students.clear()  # Clear the students dictionary, so all the previous records in dictionary will be removed for each time u creete a new file
    
    print("<< Type 'back' if want go back to the previous menu >>")
    
    while True:        
        file_name = input("Enter new file name: ") # Ask user to enter a new file name
        
        if file_name.upper() == 'BACK': # Check if user type "back" , then return to previous menu
            return 'BACK'
        
        if not file_name:  # Check if the entered file name is empty, ask user type valid input again
            print("Error: File name cannot be empty. Please enter a valid file name.")
            continue
        
        if not file_name.endswith('.txt'):  # Check if the file name does not end with '.txt'
            file_name += '.txt'              # Append '.txt' extension to the file name
        
        if not check_file_existence(file_name):  # Check if the file does not exist
            try:
                with open(file_name, 'w') as file:   # Create a new file in write mode
                    print(f"New file '{file_name}' created successfully!")
                    return file_name  # Return the file name if creation is successful
            except IOError:
                print(f"Error: Could not create file '{file_name}'. Please check permissions or disk space.")
        else:
            print(f"Error: File '{file_name}' already exists. Please enter a new file name.")

            
            
def load_records_from_file(students, file_name):  # Define a Function to load records from a file
    students.clear()  # Clear the students dictionary, so all the previous records in dictionary will be removed for each time u load data from a existing file

    print("\nLoad Record from File:")
    try:
        with open(file_name, "r") as file: # Open the file in read mode
            student_data = {}  # Temporary dictionary to store student data
            for line in file:   # Iterate through each line in the file
                line = line.strip()
                if line and not line.startswith('-'):  # Skip empty lines and lines starting with "-"
                    try:
                        key, value = line.split(":", 1)                 # Split at the first colon
                        student_data[key.strip()] = value.strip()       # Add key-value pair to student_data
                    except ValueError:
                        print(f"Issue parsing line: {line}. Skipping...")
                
                else:
                    # Create a PythonMarksheet object using data from "student_data"
                    student = PythonMarksheet(
                        student_data['Student ID'],
                        student_data.get('Student Name', ''),
                        float(student_data.get('Test 1 Mark', 0)),
                        float(student_data.get('Assignment 1 Mark', 0)),
                        float(student_data.get('Test 2 Mark', 0)),
                        float(student_data.get('Assignment 2 Mark', 0)),
                        float(student_data.get('Final Exam Mark', 0)),
                        student_data.get('Final Grade', 'NA'),  
                        float(student_data.get('Total Mark', 0))  
                    )           
                        
                    students[student_data['Student ID']] = student   # Add "student" object to "students" dictionary
                    student_data = {}  # Reset "student_data" dictionary for the next record

           
            print("Data loaded from file successfully!")

    except FileNotFoundError:  # Handle file not found error
        print("File not found. No data loaded.")

    return students  # Return the updated students dictionary
   
    
    
def display_sub_menu():   # Function to display the Student Record Management Menu 
    print("""
    >>>         *  Student Record Management  *        <<< 
-----------------------------------------------------------    
    1. Add New Student Record
    2. Display All Student Records
    3. Search Student Record by ID
    4. Update Student Record by ID
    5. Delete Student Record by ID
    6. Save Records to File
    7. Generate Grade Distribution Table
    8. Generate Grade Distribution Histogram
    9. Generate Final Marks Distribution Histogram
    10. Go back to Home Page
    11. Exit Program
-----------------------------------------------------------
    """)

def valid_marks_check(prompt):  # Function to validate/check input marks
    while True:
        value = input(prompt)  # Get input as float
        if value.upper() == 'BACK':  # Check if the user wants to go back previous menu
            return None   
        try:
            value = float(value)  # Convert the input to a float
            if 0 <= value <= 100:  # Check if the value is within the valid range
                return value  # Return the valid value
            else:
                print("Marks should be between 0 and 100.")  # Print an error message for invalid range
        except ValueError:
            print("Invalid input. Please enter a valid number.")  # Print an error message for invalid input
            

def add_new_student(students):    # Define a Function to add new student records
    print("\nAdd New Records: ")
    
    while True:
        print("<< Type 'back' if want go back to the previous menu >>")
        while True: 
            student_id = input("Enter student ID: ").upper()   # Ask user to enter student ID and change input to upper case

            if student_id == 'BACK':  # Check if the user wants to go back
                print("\n[Record not saved as you chose to go back without providing all necessary info.]")
                return  # Return to the previous menu

            # Check if the input is empty
            if not student_id:
                print("ERROR MESSAGE: Student ID cannot be empty.")
                continue

            # Check if the student ID is already in the dictionary
            if student_id in students:
                print(f"ERROR MESSAGE: The student ID {student_id} record has already been added before.")
                continue
                
            break  
            
        while True:
            student_name = input("Enter student name: ").upper()
            
            if student_name == 'BACK':  # Check if the user wants to go back
                print("\n[Record not saved as you chose to go back without providing all necessary info.]")
                return  # Return to the previous menu

            # Check if the input is empty
            if not student_name:
                print("ERROR MESSAGE: Student name cannot be empty.")
                continue
                
            break  
            
        # Input validation for test1 mark
        test1 = valid_marks_check("Enter test 1 mark: ")
        
        if test1 is None:  # Check if the user wants to go back
            print("\n[Record not saved as you chose to go back without providing all necessary info.]")
            return  # Return to the previous menu
        
        # Input validation for assignment1 mark
        assignment1 = valid_marks_check("Enter assignment 1 mark: ")
        
        if assignment1 is None:  # Check if the user wants to go back
            print("\n[Record not saved as you chose to go back without providing all necessary info.]")
            return  # Return to the previous menu

        # Input validation for test2 mark
        test2 = valid_marks_check("Enter test 2 mark: ")

        if test2 is None:  # Check if the user wants to go back
            print("\n[Record not saved as you chose to go back without providing all necessary info.]")
            return  # Return to the previous menu
        
        # Input validation for assignment2 mark
        assignment2 = valid_marks_check("Enter assignment 2 mark: ")

        if assignment2 is None:  # Check if the user wants to go back
            print("\n[Record not saved as you chose to go back without providing all necessary info.]")
            return  # Return to the previous menu
        
        final_exam = valid_marks_check("Enter final exam mark: ")

        if final_exam is None:  # Check if the user wants to go back
            print("\n[Record not saved as you chose to go back without providing all necessary info.]")
            return  # Return to the previous menu
        
        # Create an instance of PythonMarksheet
        student = PythonMarksheet(student_id, student_name, test1, assignment1, test2, assignment2, final_exam)
        
        # Calculate total marks and final grade using instance methods
        total_marks = student.calculate_total_mark()
        final_grade = student.find_final_grade()
    
        students[student_id] = student # add the 'student' object to the 'students' dictionary
        
        print("\n[The New student record has been added successfully to the current session! \nRemember to save this modification to the file if you want it also to be added permanently.]")
               
        print("\n>> Menu after added records << ")
        print("1. Continue add new records")
        print("2. Go back to Student Record Management Page")
        print("3. Exit Program")
            
        while True:    
            choice = input("Enter your choice: ")
            if choice == "1":
                break  # Continue adding new records
            elif choice == "2":
                return  # Go back to the Student Record Management Page
            elif choice == "3":
                raise ExitProgramException  # Raise exit exception to exit the whole program session
            else:
                print("Invalid choice. Please try again.")
        
        
def display_all_records(students):   # Define a Function to display all student records
    print("\nDisplay All Records:")
    
    if students:  # Check if there are any records in the students dictionary
        for student_id, student in students.items(): # Loop through each student record
            print("+" + "-"*56 + "+")  # Top border of the box
            print(f"| Student ID: {student_id.ljust(42)} |")  # Student ID line
            print("+" + "-"*56 + "+")  # Separator line
            print_student_info(student)  # Student information
            print("+" + "-"*56 + "+")  # Bottom border of the box
            print()  # Empty line for spacing
    else:
        print("No records found yet.")


def search_by_student_id(students):  # Define a Function to search for a student record by student ID
    print("\n >> Search by Student ID to Retrieve Specific Student's Marks Record <<")
    search_id = input("Enter the student ID to search (type 'back' to go back to the previous menu): ").upper()
    found = False  # Flag to track if the student record is found

    if student_id == 'BACK':
        return  # Return to previous menu
    
    if search_id in students:  # Check if the student ID exists in the students dictionary
        student = students[search_id]     # Get the student object
        print("+" + "-"*30 + "+")
        print("|" + " "*5 + "The Student's Record" + " "*5 + "|")
        print("+" + "-"*30 + "+")
        print("| Student ID: {:s}".format(student.get_student_id()).ljust(31) + "|")
        print("| Student Name: {:s}".format(student.get_student_name()).ljust(31) + "|")
        print("| Test 1 Mark: {:f}".format(student.get_test1()).ljust(31) + "|")
        print("| Assignment 1 Mark: {:f}".format(student.get_assignment1()).ljust(31) + "|")
        print("| Test 2 Mark: {:f}".format(student.get_test2()).ljust(31) + "|")
        print("| Assignment 2 Mark: {:f}".format(student.get_assignment2()).ljust(31) + "|")
        print("| Final Exam Mark: {:f}".format(student.get_final_exam()).ljust(31) + "|")
        print("| Total Mark: {:f}".format(student.get_total_mark()).ljust(31) + "|")
        print("| Final Grade: {:s}".format(student.get_final_grade()).ljust(31) + "|")
        print("+" + "-"*30 + "+")
        found = True  # Set found flag to true since the record is found
    else:
        print("Student ID not found.")

    return found   


def update_student_marks(students):  # Define a Function to update specific info for a student record by student ID
    print("\n >> Update Student's Marks Record by Student ID <<")
    student_id = input("Enter student ID to update record (type 'back' to go back to the previous menu): ").upper()
    found = False
    
    if student_id == 'BACK':
        return  # Return to previous menu if user type 'back'
        
   # Check if the record is in the dictionary
    if student_id in students:
        print("+" + "-"*50 + "+")
        print("|" + "The Student's Record :".center(50) + "|")
        print("+" + "-"*50 + "+")
        student = students[student_id]  # Get the student object from the dictionary
        print_student_info(student)     # Call function to print student information
        print("+" + "-"*50 + "+")
        print("\n>> Select what you want to update<< ") #user get to choose which info need to modify
        print("1. Student Name")
        print("2. Test 1 Mark")
        print("3. Assignment 1 Mark")
        print("4. Test 2 Mark")
        print("5. Assignment 2 Mark")
        print("6. Final Exam Mark")
        print("7. Go back to Student Record Management Page")
        print("8. Exit Program")

 
        while True:
            choice = input("Enter your choice: ")
            
            if choice == "1":  # If user chooses to update student name
                while True:
                    new_student_name = input("Enter new student name: ").upper()
                    if new_student_name.strip():  # Check if the input is not empty after stripping whitespace
                        student.set_student_name(new_student_name)  # Update student name by using set method
                        students[student_id] = student  # Update in the dictionary
                        print("\nStudent name updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")
                        break
                    else:
                        print("Student name cannot be empty.")
                        
            elif choice == "2":  # If user chooses to update test 1 mark
                new_test1 = valid_marks_check("Enter new test 1 mark: ")
                student.set_test1(new_test1)
                students[student_id] = student  # Update in the dictionary
                print("\nTest 1 mark updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")

            elif choice == "3":   # If user chooses to update assignment 1 mark
                new_assignment1 = valid_marks_check("Enter new assignment 1 mark: ")
                student.set_assignment1(new_assignment1)
                students[student_id] = student  # Update in the dictionary
                print("\nAssignment 1 mark updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")

            elif choice == "4":   # If user chooses to update test 2 mark
                new_test2 = valid_marks_check("Enter new test 2 mark: ")
                student.set_test2(new_test2)
                students[student_id] = student  # Update in the dictionary
                print("\nTest 2 mark updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")

            elif choice == "5":   # If user chooses to update assignment 2 mark
                new_assignment2 = valid_marks_check("Enter new assignment 2 mark: ")
                student.set_assignment2(new_assignment2)
                students[student_id] = student  # Update in the dictionary
                print("\nAssignment 2 mark updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")

            elif choice == "6":   # If user chooses to update final exam mark
                new_final_exam = valid_marks_check("Enter new final exam mark: ")
                student.set_final_exam(new_final_exam)
                students[student_id] = student  # Update in the dictionary
                print("\nFinal exam mark updated successfully in the current running session! \nRemember to save this modification to the file if you want it also to be updated in the file permanently.")

            elif choice == "7":
                return  # Go back to the previous menu

            elif choice == "8":
                raise ExitProgramException  # Raise custom exception to exit the whole program session

            else:
                print("Invalid choice. Please try again.")
                continue

            # Recalculate total mark and final grade after updating the record
            student.calculate_total_mark()   # Call function to recalculate total mark
            student.find_final_grade()     # Call function to find final grade
            print_student_info(student)  # Print updated student information
            print("+" + "-"*50 + "+")
            found = True   # Set found flag to true since the record is found
            break  # Break out of the loop

    else:
        print("Student ID not found.")
        found = False  # Set found flag to false

    return found  # Return the value of found flag


def delete_student_record(students):     # Define a Function to delete student record by student ID from dictionary
    print("\nDelete Student Record by Student ID:")
    student_id = input("Enter student ID to delete record (type 'back' to go back to the previous menu): ").upper()
    found = False  # Flag to track if the student record is found
    
    if student_id == 'BACK':
        return  # Return to previous menu
        
    if student_id in students:     # Check if the student ID exists in the students dictionary
        print("+" + "-"*50 + "+")
        print("|" + "The Student's Record :".center(50) + "|")
        print("+" + "-"*50 + "+")
        student = students[student_id]   # Get the student object from the dictionary
        print_student_info(student)     # Call function to print student information
        print("+" + "-"*50 + "+")
        
        while True:
            confirm = input("Are you sure you want to delete this data? (yes/no): ")  # Prompt user for confirmation

            if confirm.lower() == "yes":  # If user confirms deletion
                del students[student_id]  # Delete the student record from the dictionary
                print(f"\nStudent record with ID {student_id} deleted successfully from the current running session! \nRemember save this modification to the file if you want it also to be deleted from the file permanently.")
                break
            elif confirm.lower() == "no":  # If user cancels deletion
                print("Deletion canceled.")
                break
            else:
                print("Please enter valid option (yes/no)")
            
        found = True  # Set found flag to true since the record is found
    else:
        print("Student ID not found.")
        found = False  # Set found flag to false

    return found
        

def save_records_to_file(students, file_name):   # Define a Function to save student records to a external file
    try:
        with open(file_name, 'w') as file: # Open file in write mode
            for student_id, student in students.items():   # Iterate through student records
                # Write student information to the file
                file.write(f"Student ID: {student.get_student_id()}\n")   
                file.write(f"Student Name: {student.get_student_name()}\n")
                file.write(f"Test 1 Mark: {student.get_test1()}\n")
                file.write(f"Assignment 1 Mark: {student.get_assignment1()}\n")
                file.write(f"Test 2 Mark: {student.get_test2()}\n")
                file.write(f"Assignment 2 Mark: {student.get_assignment2()}\n")
                file.write(f"Final Exam Mark: {student.get_final_exam()}\n")
                file.write(f"Total Mark: {student.get_total_mark()}\n")
                file.write(f"Final Grade: {student.get_final_grade()}\n")
                file.write("-" * 50 + "\n")  # Write separator between student records

        print(f"\n[Records saved to file '{file_name}' successfully!]")
    except IOError:
        print(f"Error: Could not save records to file '{file_name}'. Please check permissions or disk space.")

def generate_grade_distribution_table(students):   # Define a Function to generate grade table
    print("\nGenerate Grade Distribution Table:")
    grade_distribution = {'F': 0, 'PS2': 0, 'PS1': 0, 'CR2': 0, 'CR1': 0, 'DI2': 0, 'DI1': 0, 'HD2': 0, 'HD1': 0}

    for student in students.values():
        grade_distribution[student.get_final_grade()] += 1

    print("Grade Distribution Table:")
    print("+" + "-"*14 + "+" + "-"*14 + "+")
    print("| Grade        | Frequency        |")
    print("+" + "-"*14 + "+" + "-"*14 + "+")
    
    for grade, count in grade_distribution.items():
        print(f"| {grade.ljust(12)} | {str(count).ljust(12)} |")
    print("+" + "-"*14 + "+" + "-"*14 + "+")


def generate_grade_distribution_histogram(students):  # Define a Function to generate grade histogram
    print("\nGenerate Grade Distribution Histogram:")
    grade_distribution = {'F': 0, 'PS2': 0, 'PS1': 0, 'CR2': 0, 'CR1': 0, 'DI2': 0, 'DI1': 0, 'HD2': 0, 'HD1': 0}

    for student in students.values():
        grade_distribution[student.get_final_grade()] += 1


    plt.bar(grade_distribution.keys(), grade_distribution.values(), color='#F46E8E')
    plt.xlabel('Grade Categories')
    plt.ylabel('Frequency')
    plt.title('Grade Distribution Histogram')
    plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
    plt.grid(axis='y', alpha=0.75)  # Add grid lines for better readability
    plt.show() 
    

def generate_final_mark_histogram(students):    # Define a Function to generate final marks histogram
    final_marks = [student.get_total_mark() for student in students.values()]  # Collect all final marks into a list

    max_final_mark = max(final_marks)  # Find the maximum final mark

    bins = np.arange(0, max_final_mark + 10, 10)  # Define bins with interval of 10 starting from 0 up to the maximum mark plus 10

    plt.hist(final_marks, bins=bins, edgecolor='black', color='#CF94FD', alpha=0.7)

    plt.xlabel('Final Marks Range')
    plt.ylabel('Frequency of The Final Marks')
    plt.title('Final Marks Distribution Histogram')
    plt.xticks(bins)  # Set the x-axis ticks to match the bin intervals
    plt.grid(axis='y', alpha=0.75)  # Add grid lines for better readability
    plt.show()


    

# Main function to run the program
def main():
    students = {}  # Initialize an empty "students" dictionary to store student objects
    file_name = ""
    return_to_home = False  # Flag to indicate if returning to the home page
    
    display_userManual()    # Display the user Manual 
    
    while True:
        try:
            if return_to_home:  # Check if returning to the home page
                print("Returning to the Home Page...")
                return_to_home = False  # Reset the flag after returning to the home page
                continue
            
            display_home_menu() # Call function to display the home menu
            
            while True: # Loop to handle user input for home menu
                choice = input("Enter your choice: ")
                if choice == "1": 
                    file_name = create_new_file(students)  # Call function to create a new file
                    
                    if file_name.upper() == 'BACK':
                        raise ReturnToHomePageException  # Return to previous menu

                elif choice == "2":
                    print("<< Type 'back' if want go back to the previous menu >>")
                    
                    while True:
                        file_name = input("Enter file name to load data from. Please include the file format name (e.g. '.txt'): ")
                        
                        if file_name.upper() == 'BACK':
                            raise ReturnToHomePageException  # Raise exception to return to the previous menu
                        
                        if check_file_existence(file_name):
                            students = load_records_from_file(students, file_name) # Call Load function to load record from the specified file name
                            break # Exit the loop if file loading is successful
                        else:
                            print("Error: File not found. Please enter a valid existing file name.")

                elif choice == "3":
                    while True:
                        confirm_exit = input("Are you sure you want to exit program? All unsaved records will be lost. So, pls save records to file before doing this so. (yes/no): ")
                        if confirm_exit.lower() == "yes":
                            print("Exiting the program... Thank you!")
                            return  # Exit the whole program
                        elif confirm_exit.lower() == "no":
                            raise ReturnToHomePageException  # Raise exception to return to the previous menu
                        else:
                            print("Invalid input. Please enter 'yes' or 'no'.")
                else:
                    print("Invalid choice. Please try again.")
                    continue  # Continue to the next iteration of the loop if the choice is invalid
                    
                break # Break out of the outer loop after handling the user's choice

            try:
                while True:
                    display_sub_menu() # Display the student Record Management Menu 
                    sub_choice = input("Enter your choice: ")

                    if sub_choice == "1":
                        add_new_student(students)  # Call function to add new student records to dictionary

                    elif sub_choice == "2":
                        display_all_records(students)  # Call function to display all student records

                    elif sub_choice == "3":
                        search_by_student_id(students)   # Call function to search for a student record by ID

                    elif sub_choice == "4":
                        update_student_marks(students)   # Call function to update specific info for a student record by ID

                    elif sub_choice == "5":
                        delete_student_record(students)  # Call function to delete student records from dictionary

                    elif sub_choice == "6":
                        save_records_to_file(students, file_name)   # Call function to save all updated/modified student records in dictionary to a external file

                    elif sub_choice == "7":
                        generate_grade_distribution_table(students) #call function to generate grade table

                    elif sub_choice == "8":
                        generate_grade_distribution_histogram(students) #call function to generate grade histogram

                    elif sub_choice == "9":
                        generate_final_mark_histogram(students) #call function to generate final marks histogram

                    elif sub_choice == "10":
                        confirm_homepage = input("Are you sure you want to go back to the Home Page? All unsaved records will be lost. (yes/no): ")
                        while True:
                            if confirm_homepage.lower() == "yes":
                                raise ReturnToHomePageException # return to home page
                            elif confirm_homepage.lower() == "no":
                                print("Returning to the Student Record Management Page.....")
                                break  # Break out of the loop if 'no' is entered
                            else:
                                confirm_homepage = input("Invalid input. Please enter 'yes' or 'no': ")

                    elif sub_choice == "11":
                        confirm_Exit = input("Are you sure you want to exit program? All unsaved records will be lost. So, pls save records to file before doing this so. (yes/no): ")
                        while True:     
                            if confirm_Exit.lower() == "yes":
                                raise ExitProgramException  # Raise custom exception to exit the whole program session
                            elif confirm_Exit.lower() == "no":
                                print("Returning to the Student Record Management Page.....")
                                break # go back to Student Record Management Page
                            else:
                                confirm_Exit = input("Invalid input. Please enter 'yes' or 'no'.")                  

                    else:
                        print("Invalid choice. Please try again.")
                        continue

            except ExitProgramException:
                print("Exiting the program... Thank you!")
                return # Exit the whole program

        except ReturnToHomePageException:
                    return_to_home = True  # Set the flag to indicate returning to the home page

                
if __name__ == "__main__":
    main()   # Call the main function to start the program execution
```
