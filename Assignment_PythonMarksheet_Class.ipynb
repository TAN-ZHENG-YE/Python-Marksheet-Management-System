```python
import os # Importing the os module for file operations


def check_file_existence(file_name): # Define a function to check if a file exists
    return os.path.isfile(file_name)

class ExitProgramException(Exception): # Define a custom exception for exiting the program
    pass

class ReturnToHomePageException(Exception): # Define a custom exception for returning to the home page
    pass

class PythonMarksheet: # Define a class for managing student marksheets
    def __init__(self, student_id, student_name, test1, assignment1, test2, assignment2, final_exam, total_mark=0, final_grade=None):
        # Initialize the class with student atributes
        self.__student_id = student_id
        self.__student_name = student_name
        self.__test1 = test1
        self.__assignment1 = assignment1
        self.__test2 = test2
        self.__assignment2 = assignment2
        self.__final_exam = final_exam
        self.__total_mark = total_mark   
        self.__final_grade = final_grade

        # Find total_mark and final_grade
        self.calculate_total_mark()
        self.find_final_grade()
        
    #   Define getter and setter methods for all attributes
    def get_student_id(self): 
        return self.__student_id

    def set_student_id(self, student_id): 
        self.__student_id = student_id

    def get_student_name(self):
        return self.__student_name 

    def set_student_name(self, student_name): 
        self.__student_name = student_name 
        
    def get_test1(self):     
        return self.__test1   

    def set_test1(self, test1):  
        self.__test1 = test1

    def get_assignment1(self):  
        return self.__assignment1

    def set_assignment1(self, assignment1):  
        self.__assignment1 = assignment1

    def get_test2(self):   
        return self.__test2

    def set_test2(self, test2): 
        self.__test2 = test2

    def get_assignment2(self):
        return self.__assignment2

    def set_assignment2(self, assignment2):
        self.__assignment2 = assignment2

    def get_final_exam(self):
        return self.__final_exam

    def set_final_exam(self, final_exam):
        self.__final_exam = final_exam

    def get_total_mark(self):
        return self.__total_mark

    def set_total_mark(self, total_mark):
        self.__total_mark = total_mark

    def get_final_grade(self):
        return self.__final_grade

    def set_final_grade(self, final_grade):
        self.__final_grade = final_grade

    # Define function to calculate total mark
    def calculate_total_mark(self):
        self.__total_mark = self.__test1 * 0.1 + self.__assignment1 * 0.1 + self.__test2 * 0.1 + self.__assignment2 * 0.1 + self.__final_exam * 0.6

    # Define function to find final grade based on the total marks
    def find_final_grade(self):
        total_marks = self.__total_mark
        if 0 <= total_marks < 50:
            grade = "F"
        elif 50 <= total_marks < 55:
            grade = "PS2"
        elif 55 <= total_marks < 60:
            grade = "PS1"
        elif 60 <= total_marks < 65:
            grade = "CR2"
        elif 65 <= total_marks < 70:
            grade = "CR1"
        elif 70 <= total_marks < 75:
            grade = "DI2"
        elif 75 <= total_marks < 80:
            grade = "DI1"
        elif 80 <= total_marks < 85:
            grade = "HD2"
        elif 85 <= total_marks <= 100:
            grade = "HD1"
        else:
            grade = "Invalid Grade"
        self.__final_grade = grade
        
        
        
```
